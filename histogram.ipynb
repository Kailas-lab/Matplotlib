```python
import numpy as np
import matplotlib.pyplot as plt

# Generate the data
x = np.random.normal(170, 10, 250)

# Plot a histogram
plt.hist(x, bins=10, color='green', edgecolor='black')
plt.title('Height Distribution')
plt.xlabel('Height (cm)')
plt.ylabel('Frequency')
plt.show()
```
