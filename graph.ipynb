```python
import matplotlib.pyplot as plt
import numpy as np

ypoints = np.array([3, 8, 1, 10])

plt.plot(ypoints, marker = 'o')
plt.show()

#many markers are avilable
```

```python
import matplotlib.pyplot as plt
import numpy as np

ypoints = np.array([3, 8, 1, 10])

plt.plot(ypoints, 'o:r')
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np

ypoints = np.array([23, 18, 12, 10])

plt.plot(ypoints, 'o-.', ms = 20, mec = 'r')
plt.show()

#many Line Reference is available
```

```python
import matplotlib.pyplot as plt
import numpy as np

ypoints = np.array([23, 18, 12, 10])

plt.plot(ypoints, 'o-.', ms = 20, mec = 'r',mfc = 'r')
plt.show()

#many Line Reference is available
```

```python
import matplotlib.pyplot as plt
import numpy as np

ypoints = np.array([23, 18, 12, 10])

plt.plot(ypoints, 'oc')
plt.show()

#many color dotted Reference is available
```

```python
import matplotlib.pyplot as plt
import numpy as np

ypoints = np.array([23, 18, 12, 10])

plt.plot(ypoints, 'o-.', ms = 20, mec = 'r',mfc = '#4CAF50')
plt.show()

#use headecimal values for color
```
